```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt

# NEW!
import scipy.stats as stats
```

```python
# parameters
n1 = 30   # samples in dataset 1
n2 = 40   # ...and 2
mu1 = 1   # population mean in dataset 1 (mean offset)
mu2 = 2   # population mean in dataset 2

# generate the data
data1 = mu1 + np.random.randn(n1)
data2 = mu2 + np.random.randn(n2)

# plot them
plt.plot(np.zeros(n1), data1, 'ro', markerfacecolor='w', markersize=14)
plt.plot(np.ones(n2), data2, 'bs', markerfacecolor='w', markersize=14)
plt.xlim([-1, 2])
plt.xticks([0, 1], labels=['Group 1', 'Group 2'])
plt.show()
```

```python
# t-test via stats package

# _ind = independent samples
t, p = stats.ttest_ind(data1, data2)
print(t)
print(p)
```

```python
# Common way to show t-test results in a plot
fig = plt.figure(figsize=(10, 4))
plt.rcParams.update({'font.size':20}) # Change the font size

# Add random x offset
random_num_1 = np.random.randn(n1) / 15
random_num_2 = np.random.randn(n2) / 15

plt.plot(0 + random_num_1, data1, 'ro', markerfacecolor='w', markersize=14)
plt.plot(1 + random_num_2, data2, 'bs', markerfacecolor='w', markersize=14)
plt.xlim([-1, 2])
plt.xticks([0, 1], labels=['Group 1', 'Group 2'])
# Set the title to include the t-value and p-value
plt.title(f't = {t:.2f}, p = {p:.3f}')
plt.show()
```
