```python
# import libraries
import numpy as np
```

```python
# Create a list of numbers to compute the mean and variance of
x = [1, 2, 4, 6, 5, 4, 0]
n = len(x)

# Compute the mean
mean_1 = np.mean(x)
mean_2 = np.sum(x) / n

print(mean_1)
print(mean_2)
```

```python
# Variance
var_1 = np.var(x)
var_2 = (1 / (n - 1)) * np.sum((x - mean_1) ** 2) 
print(var_1)
print(var_2)
```

```python
# Uh oh!

var_3 = np.var(x, ddof=1)

print(var_3)
print(var_2)
```

```python
# Does it matter for large N? 

N = 10000
x = np.random.randint(0, high=20, size=N)

var0 = np.var(x, ddof=0) # default
var1 = np.var(x, ddof=1) # unbiased

print(var0)
print(var1)
```
