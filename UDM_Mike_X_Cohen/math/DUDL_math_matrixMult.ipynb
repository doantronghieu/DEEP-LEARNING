```python
# import libraries
import numpy as np
import torch
```

# Using numpy

```python
# create some random matrices
A = np.random.randn(3, 4)
B = np.random.randn(4, 5)
C = np.random.randn(3, 7)

# try some multiplications...
print(np.round(a = A@B, decimals = 2)),     print('')
# print(np.round( A@C   ,2)), print(' ')
# print(np.round( B@C   ,2)), print(' ')
print(np.round(a = C.T@A, decimals = 2)),   print('')
```

# Using pytorch

```python
# Create some random matrices
A = torch.randn(3, 4)
B = torch.randn(4, 5)
C1 = np.random.randn(4, 7)
C2 = torch.tensor(data=C1, dtype=torch.float)

# Try some multiplication
print(np.round(a=A@B, decimals=2)),     print('')
# print(np.round(a=A@B.T, decimals=2))
print(np.round(a=A@C1, decimals=2)),    print('')
print(np.round(a=A@C2, decimals=2))
```
