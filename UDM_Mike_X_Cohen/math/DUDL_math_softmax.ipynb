```python
# import libraries
import numpy as np
import torch
# nn for Neural network
import torch.nn as nn
# For visualization
import matplotlib.pyplot as plt
```

```python
# "Manually" in Numpy

# The list of numbers
z = [1, 2, 3]

# Compute the Softmax output
num = np.exp(z)
den = np.sum(np.exp(z))
sigma = num / den

print(sigma)
print(np.sum(sigma))
```

```python
# Repeat with some random integers
z = np.random.randint(low=-5, high=15, size=25)
print(z)

# Compute the softmax result
num = np.exp(z)
den = np.sum(num)
sigma = num / den

# Compare
plt.plot(z, sigma, 'ko')
plt.xlabel('Original Number (z)')
plt.ylabel('Softmaxified $\sigma$')
plt.yscale('log') #
plt.title(f'$\sum\sigma$ = {np.round(np.sum(sigma), 1)}')
plt.show()
```

# Using pytorch

```python
# Create an instance of the softmax activation class
soft_func = nn.Softmax(dim=0)

# Then apply the data to that function
sigma_T = soft_func(torch.Tensor(z))

# Now we get the results
print(sigma_T)
print(torch.sum(sigma_T))
```

```python
# Show that they are the same
plt.plot(sigma, sigma_T, 'ko')
plt.xlabel('Manual Softmax')
plt.ylabel('Pytorch nn.Softmax')
plt.title(f'The two methods correlate at r = {np.corrcoef(sigma, sigma_T)[0, 1]}')
plt.show()
```
