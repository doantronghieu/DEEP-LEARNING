```python
# import libraries
import torch
import torch.nn as nn
```

# Create a maxpool instance

```python
# Create a Pool class instance with parameters
# Parameters
pool_size = 3
tride     = 3

# Create the instance
p2 = nn.MaxPool2d(kernel_size=pool_size, stride=3)
p3 = nn.MaxPool3d(kernel_size=pool_size, stride=3)

print(p2)
print(p3)
```

```python
# Create image & apply maxpooling
# Create a 2D and a 3D image
img_2D = torch.randn(1, 1, 30, 30)
img_3D = torch.randn(1, 3, 30, 30)

# All combinations of image & maxpool dimensionality
img_2D_pool_2D = p2(img_2D)
print(f'2D image, 2D maxpool: {img_2D_pool_2D.shape}')

img_3D_pool_2D = p2(img_3D)
print(f'3D image, 2D maxpool: {img_3D_pool_2D.shape}')

# img_2D_pool_3D = p3(img_2D)
# print(f'2D image, 2D maxpool: {img_2D_pool_3D.shape}')

img_3D_pool_3D = p3(img_3D)
print(f'3D image, 3D maxpool: {img_3D_pool_3D.shape}')
```

# Create a simple CNN

```python
little_net = nn.Sequential(
    # The conv-pool block
    nn.Conv2d(3, 10, 5, 3, 2),  # Convolution layer
    nn.ReLU(),                  # Activation function
    nn.AvgPool3d(3, 3),         # Average-pool
    
    # The FFN block
    nn.Flatten(),               # Vectorize to get from image to linear
    nn.Linear(588, 1),          # FC linear layer
    nn.Sigmoid()                # Output activation
)
```

```python
# Test
img = torch.rand(1, 3, 128, 128)
little_net(img)
```
