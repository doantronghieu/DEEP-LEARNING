```python
# import libraries
import torch
import torch.nn as nn
```

# Create a model and inspect its weight matrices sizes

```python
# Build a model
a_model = nn.Sequential(
    nn.Linear(10, 14),
    nn.Linear(14, 19),
    nn.Linear(19, 8)
)

print(a_model)
```

```python
# Print the sizes of the weights matrices in each layer
for i in range(len(a_model)):
    print(a_model[i].weight.shape)
```

# Build a model with inconsistent layer shapes

```python
M2 = nn.Sequential(
    nn.Linear(10,14),  # input layer
    nn.Linear(14,9),   # hidden layer
    nn.Linear(19,8),   # output layer
      )

for i in range(len(M2)):
  print( M2[i].weight.shape )
```

```python
# Build a model
M2 = nn.Sequential(
    nn.Linear(10, 14),
    nn.Linear(14, 9),
    nn.Linear(19, 8)
)

print(M2)

# Print the sizes of the weights matrices in each layer
for i in range(len(M2)):
    print(M2[i].weight.shape)
```

# Test both models with fake data

```python
# Generate the data
n_samples  = 5
n_features = 10
fake_data  = torch.randn(n_samples, n_features)

# Test the first model
print(a_model(fake_data).shape)
# Test the second model
print(M2(fake_data).shape)
```
