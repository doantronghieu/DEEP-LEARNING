```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt

# update default font size
import matplotlib
matplotlib.rcParams.update({'font.size':18})
```

# Run the experiments

```python
# Run experiment

# Number of 'trials' in the experiment
# Actual trials is 2n
N = 50
# Number of experiment repetitions
num_exps = 10000

# Initialize
accuracy  = np.zeros(num_exps)
precision = np.zeros(num_exps)
recall    = np.zeros(num_exps)
F1_score  = np.zeros(num_exps)

# Run the experiment!
for exp_i in range(num_exps):

  # Generate data
  # True positives,  aka hits
  TP  = np.random.randint(1, N)
  # False negatives, aka misses
  FN  = N - TP
  # True negatives,  aka correct rejections
  TN  = np.random.randint(1, N)
  # False positives, aka false alarms 
  FP  = N - TN

  # The four performance measures discussed in lecture  
  # Accuracy
  accuracy[exp_i]  = (TP + TN) / (2 * N)

  # Precision
  precision[exp_i]  = TP / (TP + FP)

  # Recall
  recall[exp_i]     = TP / (TP + FN)

  # Fscore
  F1_score[exp_i]   = TP / (TP + (FP + FN) / 2)
```

```python
fig,ax = plt.subplots(1,2,figsize=(18,6))

ax[0].scatter(accuracy,F1score,s=5,c=precision)
ax[0].plot([0,1],[.5,.5],'k--',linewidth=.5)
ax[0].plot([.5,.5],[0,1],'k--',linewidth=.5)
ax[0].set_xlabel('Accuracy')
ax[0].set_ylabel('F1-score')
ax[0].set_title('F1-Accuracy by precision')


ax[1].scatter(accuracy,F1score,s=5,c=recall)
ax[1].plot([0,1],[.5,.5],'k--',linewidth=.5)
ax[1].plot([.5,.5],[0,1],'k--',linewidth=.5)
ax[1].set_xlabel('Accuracy')
ax[1].set_ylabel('F1-score')
ax[1].set_title('F1-Accuracy by recall')

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

ax[0].scatter(accuracy, F1_score, s=5, c=precision)
ax[0].plot([0, 1], [0.5, 0.5], 'k--', linewidth=0.5)
ax[0].plot([0.5, 0.5], [0, 1], 'k--', linewidth=0.5)
ax[0].set_xlabel('Accuracy')
ax[0].set_ylabel('F1-score')
ax[0].set_title('F1-Accuracy by Precision')

ax[1].scatter(accuracy, F1_score, s=5, c=recall)
ax[1].plot([0, 1], [0.5, 0.5], 'k--', linewidth=0.5)
ax[1].plot([0.5, 0.5], [0, 1], 'k--', linewidth=0.5)
ax[1].set_xlabel('Accuracy')
ax[1].set_ylabel('F1-score')
ax[1].set_title('F1-Accuracy by Recall')
```
